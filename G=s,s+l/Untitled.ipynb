```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Z = np.empty((20,20))
```

```python
for i in range(20):
    for j in range(20):
        a = -1+0.1*i
        b = -1+0.1*j
        if b!=0.0 and a!=0.0:
            W = np.load('./p=1,q=2,w=100/no_new_a=%.1f,b=%.1f.npy'%(a,b))            
        elif b==0.0 and a!=0.0:
            W = np.load('./p=1,q=2,w=100/no_new_a=%.1f,b=-%.1f.npy'%(a,b))      
        elif b!=0.0 and a==0.0:
            W = np.load('./p=1,q=2,w=100/no_new_a=-%.1f,b=%.1f.npy'%(a,b))      
        elif b==0.0 and a==0.0:
            W = np.load('./p=1,q=2,w=100/no_new_a=-%.1f,b=-%.1f.npy'%(a,b))      
        
        Z[i,j]= W[1]
        #print(Z[i,j])
```

```python
X = np.arange(-1.0,1.0,0.1)
fig = plt.matshow(Z.T)
plt.xlabel(r'$ \alpha_0 $')
plt.ylabel(r'$ \beta_0 $')
xticks = np.arange(0,20,5)
xtick_labels = ['%.1f'%X[i] for i in xticks]
plt.xticks(xticks, xtick_labels)
plt.yticks(xticks, xtick_labels)
plt.gca().xaxis.tick_bottom()
plt.title(r'$\beta_\infty$')
plt.colorbar()
plt.savefig('bb.pdf', format='pdf', bbox_inches='tight')
```
