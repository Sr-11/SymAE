```python
import plotly.graph_objects as go
import pandas as pd
import numpy as np
# Read data from a csv
x, y = np.linspace(-100, 100, 100), np.linspace(-100, 100, 100)
z = np.empty((100,100))
for i in range(100):
    for j in range(100):
        z[i,j] = (x[i]/y[j]-1)**2
        if z[i,j]>10:
            z[i,j]=10
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])
fig.update_layout(width=500, height=500,
                  margin=dict(l=65, r=50, b=65, t=90))
fig.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rc('font', size=25)
fig, axs = plt.subplots(2, 2, sharex=True, sharey=True, figsize=(14,12))
fig.subplots_adjust(wspace=0, hspace=0)
#fig.subplots_adjust(top=0.9,left=0.15,bottom=0.1,right=0.95)
# N2 | [0,0] | [0,1]
#----|-------+-------
# N1 | [1,0] | [1,1] 
#    |  C1   |  C2
d=100

axs[0,0].plot(range(d),[1]*10+[0]*50+[1]*40,color='C0')
axs[1,1].plot(range(d),np.cos(np.arange(d)/d*2*np.pi),color='C0')
axs[1,0].plot(range(d),np.cos(np.arange(d)/d*2*np.pi),color='C1')
axs[1,0].plot(range(d),np.cos((25+np.arange(d))/d*2*np.pi),color='C2')


axs[1,0].set(xlabel='1st coherent state')
axs[1,1].set(xlabel='2nd coherent state')
axs[1,0].set(ylabel='1st nuisance')
axs[0,0].set(ylabel='2nd nuisance')


for ax in axs.flat:
    ax.grid(True)
```
